```python
import pandas as pd
import numpy as np
data=pd.read_csv('train.csv')
data.head()
```

```python
data = data.query("Sex=='female'")
```

```python
data.Name.tolist()
```

```python
def name(n):
    if n.find('Miss. ')>=0:
        m=n.find('Miss. ')
        n=n[m+6:]
        br=n.find(' ')
        return n[:br]
    if n.find('(')>=0:
        s=n.find('(')
        n=n[s+1:]
        se=n.find(' ')
        n=n[:se]
        if n[-1]==',':
            return n[:-1]
        else:
            return n
        
    
```

```python
data.Name=data.Name.apply(name)
data.Name=data.Name.dropna()
data.Name.value_counts()
```
